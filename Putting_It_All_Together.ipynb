**In this notebook, we will be concatenating the made csv files together**

```python
import pandas as pd
import numpy as np
```

```python
# Read the csv files:
Allen_MenFS_Elements = pd.read_csv('2024_AllenComp_MenFS_Elements.csv')
Allen_MenFS_PCS = pd.read_csv('2024_AllenComp_MenFS_PCS.csv')
Allen_WomenFS_Elements = pd.read_csv('2024_AllenComp_WomenFS_Elements.csv')
Allen_WomenFS_PCS = pd.read_csv('2024_AllenComp_WomenFS_PCS.csv')

Canada_MenFS_Elements = pd.read_csv('2024_Canada_MenFS_Elements.csv')
Canada_MenFS_PCS = pd.read_csv('2024_Canada_MenFS_PCS.csv')
Canada_WomenFS_Elements = pd.read_csv('2024_Canada_WomenFS_Elements.csv')
Canada_WomenFS_PCS = pd.read_csv('2024_Canada_WomenFS_PCS.csv')

France_MenFS_Elements = pd.read_csv('2024_France_MenFS_Elements.csv')
France_MenFS_PCS = pd.read_csv('2024_France_MenFS_PCS.csv')
France_WomenFS_Elements = pd.read_csv('2024_France_WomenFS_Elements.csv')
France_WomenFS_PCS = pd.read_csv('2024_France_WomenFS_PCS.csv')

Japan_MenFS_Elements = pd.read_csv('2024_Japan_MenFS_Elements.csv')
Japan_MenFS_PCS = pd.read_csv('2024_Japan_MenFS_PCS.csv')
Japan_WomenFS_Elements = pd.read_csv('2024_Japan_WomenFS_Elements.csv')
Japan_WomenFS_PCS = pd.read_csv('2024_Japan_WomenFS_PCS.csv')

GPFinal_MenFS_Elements = pd.read_csv('2024_GPFinal_MenFS_Elements.csv')
GPFinal_MenFS_PCS = pd.read_csv('2024_GPFinal_MenFS_PCS.csv')
GPFinal_WomenFS_Elements = pd.read_csv('2024_GPFinal_WomenFS_Elements.csv')
GPFinal_WomenFS_PCS = pd.read_csv('2024_GPFinal_WomenFS_PCS.csv')
```

```python
# Adding a country/comp and a gender column to each

# function first
def add_metadata(df, gender, competition):
    df["Gender"] = gender
    df["Competition"] = competition
    df["Season"] = "2024"
    df["Segment"] = "Free Skate"
    return df

Allen_MenFS_Elements = add_metadata(Allen_MenFS_Elements, "Men", "Allen")
Allen_MenFS_PCS = add_metadata(Allen_MenFS_PCS, "Men", "Allen")

Allen_WomenFS_Elements = add_metadata(Allen_WomenFS_Elements, "Women", "Allen")
Allen_WomenFS_PCS = add_metadata(Allen_WomenFS_PCS, "Women", "Allen")

Canada_MenFS_Elements = add_metadata(Canada_MenFS_Elements, "Men", "Canada")
Canada_MenFS_PCS = add_metadata(Canada_MenFS_PCS, "Men", "Canada")

Canada_WomenFS_Elements = add_metadata(Canada_WomenFS_Elements, "Women", "Canada")
Canada_WomenFS_PCS = add_metadata(Canada_WomenFS_PCS, "Women", "Canada")

France_MenFS_Elements = add_metadata(France_MenFS_Elements, "Men", "France")
France_MenFS_PCS = add_metadata(France_MenFS_PCS, "Men", "France")

France_WomenFS_Elements = add_metadata(France_WomenFS_Elements, "Women", "France")
France_WomenFS_PCS = add_metadata(France_WomenFS_PCS, "Women", "France")

Japan_MenFS_Elements = add_metadata(Japan_MenFS_Elements, "Men", "Japan")
Japan_MenFS_PCS = add_metadata(Japan_MenFS_PCS, "Men", "Japan")

Japan_WomenFS_Elements = add_metadata(Japan_WomenFS_Elements, "Women", "Japan")
Japan_WomenFS_PCS = add_metadata(Japan_WomenFS_PCS, "Women", "Japan")

GPFinal_MenFS_Elements = add_metadata(GPFinal_MenFS_Elements, "Men", "GPFinal")
GPFinal_MenFS_PCS = add_metadata(GPFinal_MenFS_PCS, "Men", "GPFinal")

GPFinal_WomenFS_Elements = add_metadata(GPFinal_WomenFS_Elements, "Women", "GPFinal")
GPFinal_WomenFS_PCS = add_metadata(GPFinal_WomenFS_PCS, "Women", "GPFinal")
```

```python
# Let's check to see if this was done correctly
GPFinal_WomenFS_Elements.head()
```

```python
# Now we can create the master datasets:

all_elements = pd.concat([
    Allen_MenFS_Elements,
    Allen_WomenFS_Elements,
    Canada_MenFS_Elements,
    Canada_WomenFS_Elements,
    France_MenFS_Elements,
    France_WomenFS_Elements,
    Japan_MenFS_Elements,
    Japan_WomenFS_Elements,
    GPFinal_MenFS_Elements,
    GPFinal_WomenFS_Elements
])

all_pcs = pd.concat([
    Allen_MenFS_PCS,
    Allen_WomenFS_PCS,
    Canada_MenFS_PCS,
    Canada_WomenFS_PCS,
    France_MenFS_PCS,
    France_WomenFS_PCS,
    Japan_MenFS_PCS,
    Japan_WomenFS_PCS,
    GPFinal_MenFS_PCS,
    GPFinal_WomenFS_PCS
])
```
