```python
import pdfplumber

text = ""

with pdfplumber.open("BostonMASkatingChamps.pdf") as pdf:
    for page in pdf.pages:
        text += page.extract_text() + "\n"

print(text[:2000])  # inspect first part
```

```python
# Ok now i get all scores for the skaters, lets turn this into a dataframe
import re
import pandas as pd

skater_blocks = text.split("Rank Name Nation")

all_elements = []

for block in skater_blocks[1:]:
    
    # --- Extract skater name ---
    name_match = re.search(r"\d+\s+([A-Za-z\s\-']+)\s+[A-Z]{3}", block)
    if not name_match:
        continue
    
    skater_name = name_match.group(1).strip()
    
    # --- Extract elements ---
    element_pattern = re.compile(
        r"\n\s*(\d+)\s+([A-Za-z0-9\+\*q<>]+)\s+([\d\.]+)\s+([\d\.\-]+).*?([\d\.]+)$",
        re.MULTILINE
    )
    
    for match in element_pattern.finditer(block):
        all_elements.append({
            "Skater": skater_name,
            "Element #": match.group(1),
            "Element": match.group(2),
            "Base Value": float(match.group(3)),
            "GOE": float(match.group(4)),
            "Final Score": float(match.group(5))
        })

elements_df = pd.DataFrame(all_elements)

elements_df.head()
```

```python
# Yes, this works! Note, this is only GOE, not any of the subjective based scoring
# Let's look at the parts for alysa LIU

elements_df.loc[elements_df['Skater']=='Alysa LIU',:]
```

```python
pcs_rows = []

for block in skater_blocks[1:]:
    
    # --- Extract skater name ---
    name_match = re.search(r"\d+\s+([A-Za-z\s\-']+)\s+[A-Z]{3}", block)
    if not name_match:
        continue
    
    skater_name = name_match.group(1).strip()
    
    # --- Extract PCS lines ---
    pcs_pattern = re.compile(
        r"(Composition|Presentation|Skating Skills)\s+([\d\.]+)\s+((?:\d+\.\d+\s+){8,9})(\d+\.\d+)"
    )
    
    for match in pcs_pattern.finditer(block):
        
        component = match.group(1)
        factor = float(match.group(2))
        
        # Judge scores
        judge_scores = match.group(3).strip().split()
        
        # Last value = panel average
        panel_avg = float(match.group(4))
        
        for i, score in enumerate(judge_scores):
            pcs_rows.append({
                "Skater": skater_name,
                "Component": component,
                "Judge": f"J{i+1}",
                "Score": float(score),
                "Factor": factor,
                "Panel Avg": panel_avg
            })
```

```python
pcs_df = pd.DataFrame(pcs_rows)
pcs_df.loc[pcs_df['Skater']=='Alysa LIU',:]
```

```python
# This is in long format, lets get it in wide format
pcs_wide = pcs_df.pivot_table(index=['Skater', 'Component'],columns='Judge',values='Score').reset_index()
panel_avg = pcs_df.groupby(['Skater', 'Component'])['Panel Avg'].first().reset_index()
pcs_wide = pcs_wide.merge(panel_avg, on=['Skater', 'Component'])

pcs_wide.loc[pcs_wide['Skater']=='Alysa LIU',:]
```
